## 1.1 Definitions of AI
AI can be classified along two dimensions (Human vs Rational, Thinking vs Acting).

AIMA Definitions Quadrant

- **Thinking Humanly**: Mimicking human thought (e.g., SOAR, ACT-R).  
- **Acting Humanly**: Turing Test ‚Äî imitate human behavior (e.g., chatbots).  
- **Thinking Rationally**: Using logic/laws of thought (e.g., theorem provers).  
- **Acting Rationally**: Rational agent paradigm (AIMA) ‚Äî agents that maximize performance.  


## 1.2 Formal Definition of a Rational Agent

An agent is defined by its **agent function**:  

\[ f: P^* \rightarrow A \]  

- \(P^*\) = set of all percept sequences  
- \(A\) = set of possible actions  

A rational agent selects action \(a_t\):  

\[ a_t = \arg\max_{a \in A} \; \mathbb{E}[U \mid percept\_history] \]  

Where:  
- \(U\) = performance measure (utility)  
- Expectation is over possible environment states

AI Spectrum

- **Narrow AI (Weak AI):** Domain-specific (e.g., AlphaGo, Siri).  
- **General AI (Strong AI):** Hypothetical, human-level.  
- **Superintelligence:** Beyond human capability (debated).  


## 1.4 Interdisciplinary Nature of AI
AI draws from: Mathematics, Computer Science, Cognitive Science, Neuroscience, Philosophy.

# üöÄ Lecture Notes: Section 3 ‚Äî History of AI


## 3.1 Early Foundations (1940s‚Äì1950s)
- **Alan Turing (1950):** Proposed the *Turing Test* as a criterion for intelligence.
- **Logic Theorist (1955):** First AI program by Allen Newell & Herbert Simon.
- **Dartmouth Conference (1956):** John McCarthy coined the term *Artificial Intelligence*. Considered the birth of AI as a field.

## 3.2 The Golden Years (1960s‚Äì1970s)
- Growth of symbolic AI and problem solvers (e.g., SHRDLU).
- Optimism: belief machines would match human intelligence soon.
- Limitations: computational power, unrealistic expectations.

## 3.3 The AI Winters (1970s‚Äì1980s)
- Funding cuts due to slow progress and overpromises.
- AI research stagnated.
- Term 'AI Winter' coined to describe loss of enthusiasm and investment.

## 3.4 Revival (1980s‚Äì1990s)
- **Expert Systems** (e.g., MYCIN, DENDRAL) became popular in industry.
- Development of probabilistic reasoning, Bayesian networks.
- AI applied in medicine, engineering, business.

## 3.5 Modern AI (2000s‚Äì2010s)
- Growth of **Machine Learning** with large datasets.
- Breakthroughs in computer vision, speech recognition, NLP.
- 2016: **AlphaGo** defeated world champion Lee Sedol, showcasing deep reinforcement learning.

## 3.6 Current Era (2020s)
- Rise of **Generative AI** (ChatGPT, Stable Diffusion, GPT-4, etc.).
- Widespread adoption across industries.
- Ethical debates on bias, explainability, job displacement.

# üöÄ Lecture Notes: Section 4 ‚Äî Pros and Cons of AI

Critical analysis of benefits and drawbacks with examples.

## 4.1 Pros of AI
- **Efficiency**: Automates repetitive tasks (e.g., assembly lines).
- **Accuracy**: Outperforms humans in certain tasks (e.g., medical imaging).
- **Cost Reduction**: Once deployed, AI reduces operational costs.
- **24/7 Availability**: Unlike humans, AI systems can run continuously.
- **New Discoveries**: AI accelerates scientific discovery (e.g., protein folding with AlphaFold).

## 4.2 Cons of AI
- **Bias**: If data is biased, AI amplifies it.
- **Job Displacement**: Automation threatens repetitive jobs.
- **Ethical Issues**: Surveillance, privacy violations.
- **Over-reliance**: Risk if systems fail unexpectedly.
- **Security Risks**: Adversarial attacks, deepfakes.

# üöÄ Lecture Notes: Section 6 ‚Äî Intelligence

Understanding human and machine intelligence.

## 6.1 What is Intelligence?
- **Psychological definition**: Ability to learn, reason, and adapt to new situations.
- **Computational definition**: Ability of an agent to achieve goals across a variety of environments (Legg & Hutter, 2007).
- Intelligence involves:
  - Learning from experience
  - Adapting to new situations
  - Understanding complex concepts
  - Using knowledge to manipulate environment

## 6.2 Human Intelligence
- Biological basis: neurons, brain plasticity
- Characteristics:
  - General-purpose adaptability
  - Emotional and social intelligence
  - Creativity and abstract thinking
- Limitations:
  - Cognitive biases
  - Memory constraints

## 6.3 Machine Intelligence
- Based on algorithms, data, and computation
- Characteristics:
  - Task-specific efficiency (Narrow AI)
  - Speed, accuracy, scalability
  - Lacks consciousness and common sense
- Limitations:
  - No general intelligence yet
  - Dependent on training data

## 6.5 Human vs Machine Intelligence
| Aspect | Human Intelligence | Machine Intelligence |
|--------|--------------------|----------------------|
| Learning | From experience, social context | From data, algorithms |

# üöÄ Lecture Notes: Section 8 ‚Äî Types of Intelligent Agents

AI agents can be classified into different types based on sophistication.

## 8.1 Simple Reflex Agents
- Choose actions based only on the **current percept**.
- Use condition‚Äìaction rules (if‚Äìthen).
- Example: Thermostat.
- **Limitation**: Fail in partially observable environments.

## 8.2 Model-based Reflex Agents
- Maintain an **internal state** of the environment.
- Update state based on percept history and knowledge of how the world works.
- Example: Vacuum cleaner that remembers dirt in unseen rooms.

## 8.3 Goal-based Agents
- Act to achieve explicit **goals**.
- Require **search and planning**.
- Example: GPS navigation system finding optimal route.

## 8.4 Utility-based Agents
- Use a **utility function** to evaluate outcomes.
- Make trade-offs among goals.
- Example: Self-driving car balancing speed, safety, comfort.

## 8.5 Learning Agents
- Improve performance through **experience**.
- Components: learning element, performance element, critic, problem generator.
- Example: AlphaGo improving through reinforcement learning.
