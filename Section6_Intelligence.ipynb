Understanding human and machine intelligence.

## 6.1 What is Intelligence?
- **Psychological definition**: Ability to learn, reason, and adapt to new situations.
- **Computational definition**: Ability of an agent to achieve goals across a variety of environments (Legg & Hutter, 2007).
- Intelligence involves:
  - Learning from experience
  - Adapting to new situations
  - Understanding complex concepts
  - Using knowledge to manipulate environment

## 6.2 Human Intelligence
- Biological basis: neurons, brain plasticity
- Characteristics:
  - General-purpose adaptability
  - Emotional and social intelligence
  - Creativity and abstract thinking
- Limitations:
  - Cognitive biases
  - Memory constraints

## 6.3 Machine Intelligence
- Based on algorithms, data, and computation
- Characteristics:
  - Task-specific efficiency (Narrow AI)
  - Speed, accuracy, scalability
  - Lacks consciousness and common sense
- Limitations:
  - No general intelligence yet
  - Dependent on training data

## 6.5 Human vs Machine Intelligence
| Aspect | Human Intelligence | Machine Intelligence |
|--------|--------------------|----------------------|
| Learning | From experience, social context | From data, algorithms |
| Adaptability | Highly adaptive, general-purpose | Narrow, domain-specific |
| Reasoning | Abstract, intuitive | Logical, statistical |
| Perception | Rich (vision, hearing, senses) | Limited to sensors & training |
| Creativity | Original, artistic | Pattern-based generation |
